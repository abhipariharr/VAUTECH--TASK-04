# Task 4 - Exploratory Data Analysis and Pattern Discovery

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading clean dataset
df = pd.read_csv('suicide_data_cleaned.csv')
```

```python
# Distribution of Suicide Numbers

plt.figure()
sns.histplot(df['suicides_no'])
plt.title("Distribution of Suicide Numbers")
plt.show()
```

```python
# Suicide rate by gender

sns.barplot(x = 'sex', y = 'suicides_no', data=df)
plt.title("Suicides by Gender")
plt.show()
```

```python
# Suicide Rate Over Year
year_data = df.groupby('year')['suicides_no'].sum()

year_data.plot()
plt.title("Total Suicides over Years")
plt.show()
```

```python
# Age Group Analaysis 

sns.barplot(x = 'age', y = 'suicides_no', data=df)
plt.xticks(rotation=45)
plt.title("Suicides by Age Group")
plt.show()
```

```python
# Correlation Between Features
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlatioon Heatmap")
plt.show()
```

```python
# Outlier Detection
sns.boxplot(x = df['suicides_no'])
plt.title("Outliers in Suicide Numbers")
plt.show()
```
